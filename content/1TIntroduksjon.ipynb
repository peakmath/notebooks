##Challenge 2: Finn et tall lavere enn 1 000 000 med så mange divisorer som mulig



Bruk denne koden hvis du vil:

```python
n = 20

count = 0

for i in range (1, n+1):
  if (n%i == 0):
    print(i)
    count = count + 1

print("Antall divisorer er:")
print(count)
```
